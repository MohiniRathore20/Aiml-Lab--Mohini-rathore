```
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].replace(dict(enumerate(iris.target_names)))

# Display dataset structure
print("Dataset Structure:")
print(df.info())

print("\nFirst 5 rows of the dataset:")
print(df.head())

# ------------------------------------------
# Step 2: Basic Visualizations (Matplotlib)
# ------------------------------------------

# Histogram for numerical features
df.iloc[:, :-1].hist(bins=10, figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()

# Bar Plot for categorical data
species_counts = df['species'].value_counts()
species_counts.plot(kind='bar', color=['orange', 'green', 'blue'])
plt.title('Bar Plot of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Pie Chart for categorical feature proportions
species_counts.plot(kind='pie', autopct='%1.1f%%', colors=['orange', 'green', 'blue'], startangle=140)
plt.title('Proportion of Species')
plt.ylabel('')  # Hide y-axis label for better visual
plt.show()


# Step 3: Advanced Visualizations (Seaborn)


# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.iloc[:, :-1], palette="Set2")
plt.title('Box Plot of Numerical Features')
plt.show()

# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='species', y='sepal length (cm)', data=df, palette="muted")
plt.title('Violin Plot of Sepal Length by Species')
plt.show()

# Pair Plot
sns.pairplot(df, hue='species', palette="husl", corner=True)
plt.suptitle('Pair Plot of Features', y=1.02)
plt.show()

# Heatmap of Correlation Matrix
plt.figure(figsize=(10, 6))
corr_matrix = df.iloc[:, :-1].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title('Heatmap of Feature Correlations')
plt.show()


# Conclusion


print("\nAnalysis from Visualizations:")
print("""
1. Histograms:
   - Features like `sepal width` and `petal width` show distinct distributions.
   - `Petal length` has a skewed distribution.

2. Bar Plot and Pie Chart:
   - The dataset has balanced classes: each species (setosa, versicolor, virginica) has 50 samples.

3. Box Plot and Violin Plot:
   - Significant variations in petal length and width across species.
   - Setosa species have distinct feature ranges compared to versicolor and virginica.

4. Pair Plot:
   - Strong linear relationships exist between `petal length` and `petal width`.
   - Setosa is clearly separable from the other species.

5. Heatmap:
   - High correlations between `petal length` and `petal width` (correlation â‰ˆ 0.96).
   - Weak correlation between `sepal width` and other features.
""")
```
